```python
import pandas as pd
```

```python
prodi_df = pd.read_csv("/content/dataset_cleaned.csv")
prodi_df.head()
```

```python
print("Data awal:")
print(prodi_df.head())
```

```python
prodi_df = prodi_df.loc[:, ~prodi_df.columns.str.contains('Unnamed')]
```

```python
prodi_df = prodi_df.dropna(subset=['No', 'Nama Prodi', 'Nama PT', 'Jenjang'])
```

```python
prodi_df_s1 = prodi_df[prodi_df['Jenjang'].str.strip().str.upper() == 'S1']
```

```python
prodi_df_s1.loc[:, 'Nama Prodi'] = prodi_df_s1['Nama Prodi'].str.strip()
prodi_df_s1.loc[:, 'Nama PT'] = prodi_df_s1['Nama PT'].str.strip()
prodi_df_s1.loc[:, 'Jenjang'] = prodi_df_s1['Jenjang'].str.strip()
```

```python
print("\nData setelah pembersihan (khusus S1):")
print(prodi_df_s1.head())
```

```python
prodi_df = prodi_df[(prodi_df['Jenjang'] != 'S3')]
prodi_df.head()
```
