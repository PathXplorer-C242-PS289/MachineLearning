## Data Wrangling

### Gathering Data

```python
df = pd.read_csv('data/riasec/data.csv', delimiter='\t')
df.head()
```

Since we only use users' test responses, we'll get data from columns R1-C8 for the model.

```python
df = df.loc[:, 'R1':'C8']
df.head()
```

### Assessing Data 

```python
print(df.shape)
```

48 columns, it means there will be 48 RIASEC test questions.

```python
df.info()
```

```python
df.isna().sum()
```

It appears that there are no unusualities such as missing values in our dataset. So, we'll skip the data cleaning and proceed to drop all rows that contain the value '0' in all its columns.

```python
zero_rows = df[(df == 0).all(axis=1)]
zero_rows
```

```python
df = df[~(df == 0).any(axis=1)]
```

```python
zero_rows = df[(df == 0).all(axis=1)]
zero_rows
```

```python
print(df.shape)
```

Finally, we'll save this dataframe into a new dataset that will be used as training input.

```python
df.to_csv("data/riasec/cleaned-data.csv", index=False)
print("Saved as 'cleaned-data.csv'")
```
